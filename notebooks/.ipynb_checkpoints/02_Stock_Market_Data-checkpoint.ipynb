# Stock Market Data

- In this notebook, I'll be using a module from python.org called fix_yahoo_finance that is, as its name would imply, a fix for the deprecated Yahoo Finance api. 
- I want to see if there's any relationship between mass shootings and gun company stock prices.

```python
import pandas as pd
import numpy as np
import fix_yahoo_finance as yf 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

First, I'll examine the stock performance of American Outdoor Brands (formerly Smith & Wesson), America's largest gun manuacturer.

```python
 aobc = yf.download('AOBC','2013-01-01','2018-03-31')
```

```python
aobc.head()
```

```python
aobc.shape
```

I'll go ahead and save this dataframe as a CSV file so that I don't have to keep re-downloading this data.

```python
aobc.to_csv('../data/aobc.csv')
```

Now I want to visualize this stock's performance over time:

```python
sns.set_style("white")
f, ax = plt.figure(figsize=(12,6)), plt.gca()
plt.plot(aobc.Close, lw=2, color="darkred", alpha=.7)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Price in Dollars", fontsize=18)
plt.tick_params(labelsize='x-large')
plt.title("American Outdoor Brands Stock Performance, 2013-2018", fontsize=20)
sns.despine(left=True, bottom=True)
```

Next, I want to look at Sturm Ruger & Co., the next-largest American arms manufacturer.

```python
rgr = yf.download('RGR','2013-01-01','2018-03-31')
```

```python
rgr.to_csv('../data/rgr.csv')
```

```python
sns.set_style("white")
f, ax = plt.figure(figsize=(12,6)), plt.gca()
plt.plot(rgr.Close, lw=2, color="darkred", alpha=.7)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Price in Dollars", fontsize=18)
plt.tick_params(labelsize='x-large')
plt.title("Sturm Ruger & Co. Stock Performance, 2013-2018", fontsize=20)
sns.despine(left=True, bottom=True)
```

Finally, I'll look at Vista Outdoor Inc., the third largest American company. 

```python
vsto = yf.download('VSTO', '2013-01-01','2018-03-01')
```

```python
sns.set_style("white")
f, ax = plt.figure(figsize=(12,6)), plt.gca()
plt.plot(vsto.Close, lw=2, color="darkred", alpha=.7)
plt.xlabel("Year", fontsize=18)
plt.ylabel("Price in Dollars", fontsize=18)
plt.tick_params(labelsize='x-large')
plt.title("Vista Outdoor, Inc. Stock Performance, 2013-2018", fontsize=20)
sns.despine(left=True, bottom=True)
```

I see that there is no stock information going further back than 2015. After doing a quick Google search, I've determined that this is because it formed as a spin-off of another company, ATK, and didn't exist as such before that time. For this reason, I'll leave this company out of my analysis.

Now I'll load in my 'mass' dataset in order to make some comparisons:

```python
mass = pd.read_csv('../data/mass.csv', index_col=0)
```

And plot it against the AOBC stock:

```python
from matplotlib.pyplot import text
sns.set_style("white")
f, ax = plt.figure(figsize=(20,10)), plt.gca()
plt.plot(aobc.Close, color= "black", label="AOBC Stock Price")
labels2=['Navy Yard, DC', 'Roseburg, OR', 'San Bernadino, CA', 'Orlando, FL','Las Vegas, NV', 'Sutherland Springs, TX', 'Parkland, FL']
plt.vlines(mass.index, ymin=0, ymax=aobc.Close.max(), color = "crimson", 
           label='Mass Shooting')

#this for loop will label all of my vertical lines with their corresponding shooting:
for i, row in mass.iterrows():
    text(i, 0, row['city_or_county'], rotation=90, verticalalignment='bottom', fontsize=16)
    

plt.legend(fontsize=14, loc='upper left')
plt.tick_params(labelsize='x-large')
plt.title("American Outdoor Brands Stock Price, Mass Shootings", fontsize=25)
plt.ylabel("Price per Share in Dollars", fontsize=20)
ax.yaxis.set_label_coords(-.02,.23)
plt.xlabel('Date', fontsize=20)
ax.xaxis.set_label_coords(.1, -.04)
sns.despine(left=True, bottom=True)
plt.savefig('../images/aobc_mass.png');
```

The stock price seemes to go up dramatically after two events in particular: San Bernadino and Orlando. It also begins to decline in the months between Mandalay Bay and Parkland. This might be due to companies/organizations divesting from gun stocks.

Let's see if there is a similar partern with the Ruger stock:

```python
from matplotlib.pyplot import text
f, ax = plt.figure(figsize=(20,10)), plt.gca()
plt.plot(rgr.Close, color='black', label='RGR Stock Price')
plt.vlines(mass.index, ymin=0, ymax=rgr.Close.max(), color='crimson',
           label= 'Mass Shooting')
for i, row in mass.iterrows():
    text(i, 0, row['city_or_county'], rotation=90, verticalalignment='bottom', fontsize=16)
plt.legend(fontsize=14, loc='upper left')
plt.tick_params(labelsize='x-large')
plt.title("Sturm Ruger Stock Price, Mass Shootings", fontsize=25)
plt.ylabel("Price per Share in Dollars", fontsize=20)
ax.yaxis.set_label_coords(-.02,.23)
plt.xlabel('Date', fontsize=20)
ax.xaxis.set_label_coords(.1, -.04)
sns.despine(left=True, bottom=True)
plt.savefig('../images/rgr_mass.png');
```

Ruger stock does very well after Navy yard, San Bernadino and Orlando and holds steady in the interim between Mandalay Bay and Parkland. If there is a correlation at all, it's certainly different from the pattern seen with AOBC.
