# Initial Preprocessing

- In this notebook, I'll read in my main dataset, clean it, and perform some initial EDA
- It would also be helpful to make some plots of overall national trends

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

The following csv file, downloaded from Kaggle user James Ko, will be my main dataset throughout this project. It is comprised of data scraped from gunviolencearchive.org. 

```python
guns = pd.read_csv('../data/guns.csv')
```

I first want to call .shape() on this dataframe to get a sense of how many rows and columns I'll be working with.

```python
guns.shape
```

And then I'll look at the head and tail of my dataset to get a sense of how it's formatted, as well as to see the dates it ranges.

```python
guns.head()
```

```python
guns.tail()
```

I can see that my data spans the timeframe of 1/1/13 to 3/31/2018.

Now I want to see which of my columns contains null values:

```python
guns.isnull().sum()
```

Most of these columns I won't needâ€“I'm primarily interested in dates, cities/states and number killed. I'll drop the 'incident id' column to begin with, because I definitely won't be using that.

```python
guns.drop("incident_id", axis=1, inplace=True)
```

From the dataset's documentation, I know that the Mandalay Bay mass shooting is not included in the dataset, bu i'll double check anyway:

```python
nevada = guns[guns['state']=='Nevada']
```

```python
nevada[nevada['date'] =='2017-10-01']
```

It's notably absent from the Kaggle data. This is terrible, because it's a historic data point that's simply too important to ignore. However, I can pretty quickly add it in myself:

```python
guns = guns.append({'date':'2017-10-01', 'state':'Nevada', 'city_or_county': 'Las Vegas', 
             'address':'3950 S Las Vegas Blvd', 'n_killed': 59, 'n_injured': 441
             }, ignore_index=True)
```

Now I'm going to set the 'date' column to the 'datetime' datatype and set that column as my index. This will be essentially for basically everything I'm going to do in this project.

```python
guns.date = pd.to_datetime(guns.date)
guns.set_index('date', inplace=True)
guns.sort_index(inplace=True)
guns.head(1)
```

Now I'll make sure that the Vegas shooting was added in:

```python
mask = guns['city_or_county'] == 'Las Vegas'
mask &= guns.index == '2017-10-01'
guns[mask]
```

I'm definitely going to want to save a copy of my cleaned/amended dataframe to use in future notebooks:

```python
guns.to_csv('../data/guns2.csv')
```

Let's get some quick totals by year:

```python
yearly = pd.DataFrame(guns.groupby([lambda x: x.year]).n_killed.sum())
```

```python
yearly
```

In order to examine yearly totals, I need to drop years 2013 and 2018 because I don't have complete data from those years.

```python
yearly.drop(index = [2013, 2018], inplace=True)
```

And now I'll go ahead and plot this:

```python
sns.set_style("white")
f, ax = plt.figure(figsize=(12,6)), plt.gca()
plt.plot(yearly.index, yearly.n_killed, lw=5, color="darkred", alpha=.7)
plt.xlabel("Year", fontsize=18)
plt.ylabel("# Killed", fontsize=18)
plt.title("US Gun Deaths by Year", fontsize=20)
ax.set_xticks([2014, 2015, 2016, 2017], )
ax.tick_params(labelsize='x-large')
sns.despine(left=True, bottom=True)
plt.savefig('../images/yearly_gd.png')
```

This plot shows an increase in gun violence over the years in my data, but I need to calculate per capita gun violence to get an accurate picture.

Since I'm only looking at four years, I'll enter population totals manually from the US Census Bureau's website (faster, in this case, than scraping).

```python
yearly['pop'] = [318622525, 321039839, 323405935, 325719178]
yearly['pc_killed'] = (yearly['n_killed'] / yearly['pop'])*100000
```

```python
yearly
```

```python
sns.set_style("white")
f, ax = plt.figure(figsize=(12,6)), plt.gca()
plt.plot(yearly.index, yearly.pc_killed, lw=5, color="darkred", alpha=.7)
plt.xlabel("Year", fontsize=18)
plt.ylabel("# Killed per 100,000", fontsize=18)
plt.title("US Per Capita Gun Deaths by Year", fontsize=20)
ax.set_xticks([2014, 2015, 2016, 2017], )
ax.tick_params(labelsize='x-large')
sns.despine(left=True, bottom=True)
plt.savefig('../images/yearly_pc_gd.png')
```

Unmistakeably, gun deaths are trending up per capita nationwide. 

Now I want to look at the City of Baltimore in particular. 

```python
bmore = guns[guns.city_or_county=='Baltimore']
```

```python
bmore.n_killed.sum()
```

There were over a thousand gun murders in Baltimore for the timespan covered in my dataset. The actual number is even higher, since 2013 wasn't fully documented and the data from 2018 is several months old. Let's look at yearly totals:

```python
bmore_yearly = pd.DataFrame(bmore.groupby([lambda x: x.year]).n_killed.sum())
```

```python
bmore_yearly
```

```python
bmore_yearly.drop([2013, 2018], axis=0, inplace=True)
```

```python
bmore_yearly['pop'] = [624062, 623031, 616958, 611648]
```

```python
bmore_yearly['pc_killed'] = (bmore_yearly['n_killed']/bmore_yearly['pop']) * 100000
```

```python
import seaborn as sns
sns.set_style("white")
f, ax = plt.figure(figsize=(12,6)), plt.gca()
plt.plot(bmore_yearly.index, bmore_yearly.pc_killed, lw=5, color="darkred", alpha=.7)
plt.xlabel("Year", fontsize=18)
plt.ylabel("# Killed per 100,000", fontsize=18)
plt.title("Baltimore Gun Deaths by Year", fontsize=20)
ax.set_xticks([2014, 2015, 2016, 2017], )
ax.tick_params(labelsize='x-large')
sns.despine(left=True, bottom=True)
plt.savefig('../images/bmore_yearly.png')
```

This line plot shows how gun murders almost doubled in Baltimore between 2014 and 2017 (despite a shrinking population).
