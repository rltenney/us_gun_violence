## Manually appending the 2017 Las Vegas mass shooting:

```python
guns = guns.append({'date':'2017-10-01', 'state':'Nevada', 'city_or_county': 'Las Vegas', 
             'address':'3950 S Las Vegas Blvd', 'n_killed': 59, 'n_injured': 441
             }, ignore_index=True)
```

```python
guns.date = pd.to_datetime(guns.date)
guns.set_index('date', inplace=True)
guns.sort_index(inplace=True)
guns.head(1)
```

```python
mask = guns['city_or_county'] == 'Las Vegas'
mask &= guns.index == '2017-10-01'
guns[mask]
```

```python
guns.to_csv('./data/guns2.csv')
```

```python
yearly = pd.DataFrame(guns.groupby([lambda x: x.year]).n_killed.sum())
```

```python
yearly
```

#### In order to examine yearly totals, I need to drop years 2013 and 2018 because I don't have complete data from those years.

```python
yearly.drop(index = [2013, 2018], inplace=True)
```

```python
plt.plot(yearly.index, yearly.n_killed)
```

## This plot shows an increase in gun violence over the years in my data, but i need to calculate per capita gun violence to get an accurate picture.
### Since I'm only looking at four years, I'll enter population totals manually from the US Census Bureau's website.

```python
yearly['pop'] = [318622525, 321039839, 323405935, 325719178]
yearly['pc_killed'] = (yearly['n_killed'] / yearly['pop'])*100000
```

```python
yearly
```

```python
import seaborn as sns
sns.set_style("white")
f, ax = plt.figure(figsize=(12,6)), plt.gca()
plt.plot(yearly.index, yearly.pc_killed, lw=3, color="crimson")
plt.xlabel("Year", fontsize=18)
plt.ylabel("# Killed per 100,000", fontsize=18)
plt.title("US Per Capita Gun Deaths by Year", fontsize=20)
ax.set_xticks([2014, 2015, 2016, 2017], )
ax.tick_params(labelsize='x-large')
sns.despine(left=True, bottom=True)
plt.savefig('./images/yearly_gd.png')
```

### Unmistakeably, gun deaths are up per capita nationwide.
