```python
guns = pd.read_csv('./data/guns2.csv')
```

```python
guns2 = guns.set_index('date')
```

```python
guns2.columns
```

```python
guns2 = guns2[['n_killed', 'state', 'city_or_county']]
```

```python
gd_state = guns2.groupby('state')[['n_killed']].sum()
gd_state.head()
```

```python
gd_state.to_csv('./data/gd_state.csv')
```

```python
sorted_gd = gd_state.n_killed.sort_values()
```

```python
pd.DataFrame(sorted_gd).to_csv('./data/gd_state_sorted.csv')
```

```python
fig, ax = plt.figure(figsize=(15,10)), plt.gca()
sns.set_style("white")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"
plt.barh(sorted_gd.index, sorted_gd, color = 'black', alpha =.7, align='center', lw=.5)
plt.yticks(fontsize=13)
plt.xticks(fontsize=16)
plt.title("Gun Deaths by State", fontsize=30)
plt.tight_layout()
plt.savefig('./images/total_gun_deaths_by_state.png')
```

```python
gd_by_date = guns.groupby('date')[['n_killed']].max()
```

```python
gd_by_date.tail()
```

```python
gd_by_date[gd_by_date['n_killed'] >= 10]
```

```python
aobc.loc['2017-09-28':'2017-10-07','Close']
```

```python
sns.set_style("white")
f, ax = plt.figure(figsize=(20,10)), plt.gca()
plt.plot(aobc.Close, color= "black", label="AOBC Stock Price")
plt.vlines(gd_by_date[gd_by_date['n_killed'] >= 10].index, ymin=0, ymax=aobc.Close.max(), color = "crimson", 
           label="Mass Shooting")
plt.legend(fontsize=14, loc='upper left')
plt.tick_params(labelsize='x-large')
plt.title("American Outdoor Brands Stock Price, Mass Shootings", fontsize=25)
plt.ylabel("Price per Share in Dollars", fontsize=20)
ax.yaxis.set_label_coords(-.02,.23)
plt.xlabel('Date', fontsize=20)
ax.xaxis.set_label_coords(.1, -.04)
sns.despine(left=True, bottom=True)
plt.savefig('./images/aobc_mass.png');
```

```python
f, ax = plt.figure(figsize=(20,10)), plt.gca()
plt.plot(rgr.Close, color='black', label='RGR Stock Price')
plt.vlines(gd_by_date[gd_by_date['n_killed'] >= 10].index, ymin=0, ymax=rgr.Close.max(), color='crimson',
           label= 'Mass Shooting' )
plt.legend(fontsize=14, loc='upper left')
plt.tick_params(labelsize='x-large')
plt.title("Sturm Ruger Stock Price, Mass Shootings", fontsize=25)
plt.ylabel("Price per Share in Dollars", fontsize=20)
ax.yaxis.set_label_coords(-.02,.23)
plt.xlabel('Date', fontsize=20)
ax.xaxis.set_label_coords(.1, -.04)
sns.despine(left=True, bottom=True)
plt.savefig('./images/rgr_mass.png');
```
