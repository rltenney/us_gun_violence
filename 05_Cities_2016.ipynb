```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## In this notebook, I'm going to repeat the steps I took in the previous notebook, this time looking at 2016 data.

```python
df = pd.read_csv('../guns2_clean.csv', index_col = 0)
```

```python
df.head()
```

```python
df_2016 = df.loc['2016-01-01':'2016-12-31']
```

```python
pd.options.mode.chained_assignment = None
```

```python
df_2016['city'] = df_2016['city'] + ", " + df_2016['state']
```

```python
df_2016.drop(['state'], axis=1, inplace=True)
```

```python
df_2016.head()
```

```python
cities_2016 = pd.DataFrame(df_2016.groupby(['city']).n_killed.sum().sort_values(ascending=False))
```

```python
cities_2016.head()
```

## Loading in Census Bureau csv: 

```python
pop = pd.read_csv('pop.csv', header=1, index_col='Rank')
```

```python
pop['Geography.2'].replace('city.+', "", regex=True, inplace=True)
```

```python
pop['Geography.1'].replace('^(.+?),', "", regex=True, inplace=True)
```

```python
pop = pop[['Geography.2', 'Geography.1', 'Population Estimate (as of July 1) - 2016']]
```

```python
pop.rename(columns= {'Geography.2':'city', 'Geography.1':'state', 'Population Estimate (as of July 1) - 2016':'2016_pop' }, inplace=True)
```

```python
pop['city'] = pop['city'].str.strip()
```

```python
pop['state'] = pop['state'].str.strip()
```

```python
pop['city'] = pop['city'] + ", " + pop['state']
```

```python
pop.set_index(['city'], inplace=True)
```

```python
pop.drop('state', axis= 1, inplace=True)
```

```python
pop.head()
```

## Merging population and gun death dataframes:

```python
cities_2016 = pd.merge(pop, cities_2016, left_index=True, right_index=True)
```

```python
cities_2016.sort_values('n_killed', ascending=False).head(20)
```

```python
cities_2016['pc_killed'] = (cities_2016.n_killed / cities_2016['2016_pop']) * 100000
```

```python
top_cities_2016 = cities_2016.sort_values('pc_killed', ascending=False).head(20)
top_cities_2016
```

```python
top_large_cities_2016 = cities_2016[cities_2016['2016_pop']>=500000].sort_values('pc_killed', ascending=False).head(20)
top_large_cities_2016
```

```python
sns.set(style="white")

f, ax = plt.subplots(figsize=(10, 8))
colors = ['orange' if (x=='Baltimore, Maryland' or x == 'Gary, Indiana') else 'grey' for x in top_cities_2016.index]
sns.barplot(x=top_cities_2016.pc_killed, 
            y=top_cities_2016.index,
            data=top_cities_2016, palette=colors, alpha=.7)
sns.set(font_scale=2)
ax.set(xlabel="Gun Deaths per 100k",
       ylabel="City",
       title="US Cities by Per Capita Gun Deaths 2016")
sns.despine(left=True, bottom=True)
```

```python
sns.set(style="white")

f, ax = plt.subplots(figsize=(10, 8))
colors = ['orange' if (x=='Baltimore, Maryland' or x == 'Washington, District of Columbia') else 'grey' for x in top_large_cities_2016.index]
sns.barplot(x=top_large_cities_2016.pc_killed, 
            y=top_large_cities_2016.index,
            data=top_large_cities_2016, palette=colors, alpha=.7)
sns.set(font_scale=2)
ax.set(xlabel="Gun Deaths per 100k",
       ylabel="City",
       title="Large US Cities by Per Capita Gun Deaths 2016")
sns.despine(left=True, bottom=True)
```
