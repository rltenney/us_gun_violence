## Loading in Census Bureau csv file:

```python
pop = pd.read_csv('pop.csv', header=1, index_col='Rank')
```

```python
pop.head()
```

### Using a bit of regex to clean up my city names:

```python
pop['Geography.2'].replace('city.+', "", regex=True, inplace=True)
```

```python
pop['Geography.1'].replace('^(.+?),', "", regex=True, inplace=True)
```

```python
pop = pop[['Geography.2', 'Geography.1', 'Population Estimate (as of July 1) - 2017']]
```

```python
pop.rename(columns= {'Geography.2':'city', 'Geography.1':'state', 'Population Estimate (as of July 1) - 2017':'2017_pop' }, inplace=True)
```

```python
pop['city'] = pop['city'].str.strip()
```

```python
pop['state'] = pop['state'].str.strip()
```

```python
pop['city'] = pop['city'] + ", " + pop['state']
```

```python
pop.set_index(['city'], inplace=True)
```

```python
pop.drop('state', axis= 1, inplace=True)
```

```python
pop.tail()
```

## Merging population and gun death dataframes:

```python
cities_2017 = pd.merge(pop, cities_2017, left_index=True, right_index=True)
```

```python
cities_2017.loc['Los Angeles, California',:]
```

```python
cities_2017.sort_values('n_killed', ascending=False)
```

```python
cities_2017['pc_killed'] = (cities_2017.n_killed / cities_2017['2017_pop']) * 100000
```

```python
len(cities_2017[cities_2017['2017_pop']>= 100000].sort_values('pc_killed', ascending=False))
```

```python
top_cities_2017 = cities_2017.sort_values('pc_killed', ascending=False).head(20)
top_cities_2017
```

```python
top_large_cities_2017 = cities_2017[cities_2017['2017_pop']>=500000].sort_values('pc_killed', ascending=False).head(20)
top_large_cities_2017
```

```python
top_large_cities_2017 = top_large_cities_2017.reset_index()
```

```python
top_large_cities_2017['city'], top_large_cities_2017['state'] = top_large_cities_2017.city.str.split(', ', 1).str
```

```python
top_large_cities_2017.head()
```

```python
top_large_cities_2017.set_index('city', inplace=True)
```

```python
top_large_cities_2017.to_csv('top_large_cities.csv')
```

```python
sns.set(style="white")

f, ax = plt.subplots(figsize=(10, 8))
colors = ['orange' if (x=='Baltimore, Maryland' or x == 'Wilmington, Delaware') else 'grey' for x in top_cities_2017.index]
sns.barplot(x=top_cities_2017.pc_killed, 
            y=top_cities_2017.index,
            data=top_cities_2017, palette=colors, alpha=.7)


ax.set_title("US Cities by Per Capita Gun Deaths 2017", fontsize=20)
ax.set_xlabel('Gun Deaths per 100k',fontsize=14)
ax.set_ylabel("City", fontsize=14)
       
sns.despine(left=True, bottom=True)
```

```python
sns.set(style="white")

f, ax = plt.subplots(figsize=(10, 8))
colors = ['orange' if (x=='Baltimore' or x == 'Washington') else 'grey' for x in top_large_cities_2017.index]
sns.barplot(x=top_large_cities_2017.pc_killed, 
            y=top_large_cities_2017.index,
            data=top_large_cities_2017, palette=colors, alpha=.7)

ax.set_title("Large US Cities by Per Capita Gun Deaths 2017", fontsize=20)
ax.set_xlabel('Gun Deaths per 100k',fontsize=14)
ax.set_ylabel("City", fontsize=14)

sns.despine(left=True, bottom=True)
```
